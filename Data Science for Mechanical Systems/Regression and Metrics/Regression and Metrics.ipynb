All codes used for this problem are below

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("simple_linear_regression.csv",header=None)
data.columns = ['x','y']
data['x2'] = [x**2 for x in data.x]
```

```python
#for y = β0 +β1x
ols = smf.ols(formula='y~x', data=data)
ols_result = ols.fit()
ols_result.summary()
```

```python
#print("Prediction of x=10 for the fitted model : {}",ols_result.predict([10]))
```

```python
#for y = β0+β1x+β2x2 
ols = smf.ols(formula='y~x+x2', data=data)
ols_result = ols.fit()
ols_result.summary()
```
